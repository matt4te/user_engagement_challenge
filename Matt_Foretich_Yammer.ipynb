# Working Hypotheses

In order to guide the effort in determining the decline in user engagement, it would be useful to start with a couple of hypotheses based on the data available. I think that these sound fairly reasonable:

##### (1) There has been a decline in user signups. Acknowleding that some churn from the app is likely inevitable, this would result in an overall decline in the number of engaged users.

##### (2) There has been declining user engagement in particular countries (indicated by location in the events data) or regions (we might use a user's preferred language as a proxy)

##### (3) There has been declining user engagement by particular device users (possibly Yammer has lagged behind in their support for either PC or Mac or Mobile)

##### (4) Certain companies have stopped using Yammer 

##### (5) Companies have begun to migrate email activities away from Yammer. Thus, users who used to engage with Yammer for this particular reason only are less likely to engage during any week

&nbsp;

With this ideas in mind, the effort to help out the head of product can be a bit more focused.

# Exploring the Data

&nbsp;

```python
# load altair and its renderer to visualize the data
import altair as alt
alt.renderers.enable('notebook')
alt.data_transformers.enable('json')
```

```python
# check data types of users dataframe

users.user_id = users.user_id.astype('category')
users.created_at = pd.to_datetime(users.created_at, errors='coerce') # coerce NaN to NaT
users.company_id = users.company_id.astype('category')
users.activated_at = pd.to_datetime(users.activated_at, errors='coerce') # coerce NaN to NaT

print(users.dtypes)
```

```python
# and lets create year/month identifiers for signup and activation dates

users['month_created'] = users['created_at'].map(lambda x: 12*x.year + x.month) # *12 creates separation between the years
users['month_activated'] = users['activated_at'].map(lambda x: 12*x.year + x.month)
```

```python
# First, lets look at user signups over time

create_hist = alt.Chart(users).mark_bar().encode(
    alt.X('month_created', axis=alt.Axis(title='Month Account Created', labels=False)),
    alt.Y('count()')
).properties(title='Number of Users Signing up to Yammer, by month')

create_hist
```

```python
# And the number of users activated accounts over time

activate_hist = alt.Chart(users).mark_bar().encode(
    alt.X('month_activated', axis=alt.Axis(title='Month Account Activated', labels=False)),
    alt.Y('count()')
).properties(title='Number of users activating Yammer accounts, by month')

activate_hist
```

&nbsp;

### Conclusion: It seems that there hasn't been a decline in user sign-up or activation. In fact, there has been a steady increase

&nbsp;

#### Let's continue by looking at the events data

&nbsp;

```python
# We can now safely exclude all the events related to sign-up activities

events = events[events.event_type=='engagement']
```

```python
events.head()
```

```python
# we only have the 4 most recent months of events, so lets create a column to indicate the month of each event

events.occurred_at = pd.to_datetime(events.occurred_at)
events['event_month'] = events.occurred_at.dt.month
```

```python
# and visualize how the most recent month's engagement compares across different event types

event_hist = alt.Chart(events).mark_bar().encode(
    alt.X('event_month', axis=alt.Axis(title='Month Event Occured', labels=False)),
    alt.Y('count(event_name)'),
    row='event_name',
    color=alt.condition(
        alt.datum.event_month == 8,  # If the year is 1970 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(title='Number of engagement activities, by month')

event_hist
```

### Conclusion: This, more or less, confirms what we already knew - there has been a substantial decrease in activity. It is particularly important that the number of logins and homepage visits has declined, which reflects a drop in the number of active users, as these are the first steps in engaging with Yammer

&nbsp;

```python
# Lets focus on homepage visits, since this is the first step in user engagement

homepage = events[events.event_name == 'home_page']
homepage.head()
```

```python
# and look towards our second hypothesis - that engagement might be declining in certain countries

country_hist = alt.Chart(homepage).mark_bar().encode(
    alt.X('event_month', axis=alt.Axis(title='Month Event Occured', labels=False)),
    alt.Y('count(event_name)'),
    row='location',
    color=alt.condition(
        alt.datum.event_month == 8,  # If the year is 1970 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(title='Number of homepage visits, by country')

country_hist
```

### Conclusion: Though there have been slight engagemement increases in some countries (e.g., UAE, Turkey, Spain, Argentina), there has been a decrease in activity in the majority of countries, including Yammer's most valuable markets (e.g., US, France, Germany, Japan). It seems unlikely that the drop in user engagement can be attributed to country differences.

&nbsp;

```python
# so lets look towards our third hypothesis - that engagement might be declining in certain countries

device_hist = alt.Chart(homepage).mark_bar().encode(
    alt.X('event_month', axis=alt.Axis(title='Month Event Occured', labels=False)),
    alt.Y('count(event_name)'),
    row='device',
    color=alt.condition(
        alt.datum.event_month == 8,  # If the year is 1970 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(title='Number of homepage visits, by device type')

device_hist
```

### CONCLUSION: It appears that there has been a big drop in the number of homepage visits on mobile devices, but its hard to tell with so many categories. We should group the devices by mobile or not

&nbsp;

```python
# lets group the devices by whether theyre mobile or not

import numpy as np

cpu = ['macbook pro', 'lenovo thinkpad', 'macbook air', 'dell inspiron notebook', 'dell inspiron desktop', 'asus chromebook', 
      'acer inspire notebook', 'hp pavilion desktop', 'acer inspire desktop', 'mac mini', 'windows surface']

mobile = homepage.copy()
mobile['mobile'] = np.nan

for i in range(len(mobile)):
    if mobile.loc[mobile.index[i], 'device'] in cpu:
        mobile.loc[mobile.index[i], 'mobile']='cpu'
    else: mobile.loc[mobile.index[i], 'mobile']='mobile'
```

```python


mobile_hist = alt.Chart(mobile).mark_bar().encode(
    alt.X('event_month', axis=alt.Axis(title='Month Event Occured', labels=False)),
    alt.Y('count(event_name)'),
    row='mobile',
    color=alt.condition(
        alt.datum.event_month == 8,  # If the year is 1970 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(title='Number of homepage visits, by device type')

mobile_hist
```

### Conclusion: While it seems that most of the homepage view declines are coming from a drop in mobile views, there is still a decline for cpu users, so something else must be going on in addition.

&nbsp;

## Moving Forward

Now we have an intution for what is going on: there is a general decline in log-on activity, and this may have something to do with a decline in mobile use. However, what the product head wants to know is why are we losing USERS, not homepage visits, so we need to join the engagement data with the user data

&nbsp;

```python
# lets include company data with our events by joining with the user table

mobile.user_id = mobile.user_id.astype('category')
comps = pd.merge(mobile, users, on='user_id', how='left')
```

```python
# create a dataframe that has the number of companies represented in the data for each month

month_comps = pd.DataFrame(comps.groupby(['event_month', 'company_id']).count().groupby('event_month').count())
```

```python
comp_hist = alt.Chart(month_comps.reset_index()).mark_bar().encode(
    alt.X('event_month', axis=alt.Axis(title='Month', labels=False)),
    alt.Y('user_id', axis=alt.Axis(title='Number of Companies')),
    color=alt.condition(
        alt.datum.event_month == 8,  # If the year is 1970 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(title='Number of companies using Yammer, by month')

comp_hist
```

### Conclusion: It appears there has been a pretty substantial decline in the number of companies using Yammer between July and August - this would substantially reduce the number of engaged users. 

&nbsp;

```python
# Next, lets look at the email data to see if trends in emailing activity have changed in august

emails.occurred_at = pd.to_datetime(emails.occurred_at)
emails['email_month'] = emails.occurred_at.dt.month

emails.head()
```

```python
emails.action.value_counts()

email_hist = alt.Chart(emails).mark_bar().encode(
    alt.X('email_month', axis=alt.Axis(title='Month Event Occured', labels=False)),
    alt.Y('count(action)'),
    row='action',
    color=alt.condition(
        alt.datum.email_month == 8,  # If the year is 1970 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(title='Number email activities per month')

email_hist
```

```python
# calculate the number of unique users involved in email activities each month

user_emails = pd.DataFrame(emails.groupby(['email_month','user_id']).count().groupby('email_month').count())
```

```python
euser_hist = alt.Chart(user_emails.reset_index()).mark_bar().encode(
    alt.X('email_month', axis=alt.Axis(title='Month', labels=False)),
    alt.Y('user_type', axis=alt.Axis(title='Number of Unique Email Users')),
    color=alt.condition(
        alt.datum.email_month == 8,  # If the year is 1970 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(title='Number of unique users using email functionality, by month')

euser_hist
```

### Conclusion: There are actually more unique users utilizing the email functionality of Yammer recently. Thus, migration of email activities away from Yammer is not contributing to the decline in engagement.
